```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando 50 números aleatórios entre 100.000 e 500.000
np.random.seed(42)
values = np.random.randint(low=100000, high=500000, size=50)

# Plotando o gráfico
plt.plot(values, color='blue')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('50 pontos aleatórios')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Criando o modelo de regressão linear
model = LinearRegression()

# Treinando o modelo com os dados existentes
X = np.arange(len(values)).reshape(-1, 1)
y = values.reshape(-1, 1)
model.fit(X, y)

# Fazendo a previsão dos próximos 3 pontos
X_pred = np.array([len(values), len(values)+1, len(values)+2]).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Plotando os dados e a previsão
plt.plot(values, color='blue', label='Dados existentes')
plt.plot(np.concatenate([X[-1:], X_pred]), np.concatenate([y[-1:], y_pred]), color='red', label='Previsão')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('50 pontos aleatórios e previsão com 3 pontos à frente')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Gerando 50 números aleatórios entre 100.000 e 500.000
np.random.seed(42)
values = np.random.randint(low=100000, high=500000, size=50)

# Separando os dados em treinamento e teste
X_train = np.arange(40).reshape(-1, 1)
y_train = values[:40].reshape(-1, 1)
X_test = np.arange(40, 50).reshape(-1, 1)
y_test = values[40:].reshape(-1, 1)

# Criando o modelo de regressão linear e treinando com os dados de treinamento
model = LinearRegression()
model.fit(X_train, y_train)

# Fazendo a previsão dos próximos 3 pontos com os dados de teste
X_pred = np.array([len(values)-3, len(values)-2, len(values)-1]).reshape(-1, 1)
y_pred = model.predict(X_pred)

# Plotando os dados, a previsão e o teste
plt.plot(values, color='blue', label='Dados')
plt.plot(np.concatenate([X_train, X_test]), np.concatenate([y_train, y_test]), color='green', label='Dados de treinamento/teste')
plt.plot(np.concatenate([X_train, X_test[-1:]]), np.concatenate([model.predict(X_train), y_test[:1]]), color='orange', label='Previsão no teste')
plt.plot(X_pred, y_pred, 'o', color='red', label='Próximos 3 pontos previstos')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.title('50 pontos aleatórios com previsão e teste')
plt.legend()
plt.show()
```
